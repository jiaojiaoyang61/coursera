scrape the table from wiki page. Then delete the 'not assigned' rows, group the table by 'Postal Code'

```python
import pandas as pd

dfs = pd.read_html('https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M') 
```

```python
df = dfs[0]
df.columns = ['Postal Code','Borough','Neighborhood']
```

```python
df = df.drop(index=0) # delete one row
```

```python
df_new = df[df.Borough!='Not assigned']
df_new
```

```python
df_group = df_new.groupby('Postal Code')
```
